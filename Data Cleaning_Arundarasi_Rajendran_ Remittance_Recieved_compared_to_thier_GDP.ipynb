• Fill missing values with forward fill

```python
df1 = df.fillna(method='ffill')
df1.isnull().sum()
```

• Drop missing values

```python
df2 = df1.dropna()
df2.isna().sum()
```

```python
df3 = df2.drop(columns=['Country Code'])
```
