H6: Countries in certain regions are more likely to be considered as high income group

```python
df['new_income_group'].unique()
```

```python
income_mapping = {'Low income':1, 'Lower middle income':2,'Upper middle income':3,'High income':4}

df.head()
```

```python
df['new_income_group']=df['new_income_group'].map(income_mapping)
```

```python
region = df['new_region']
avg_income = df.groupby('new_region')['new_income_group'].mean()
avg_income
```

```python
contingency_tab = pd.DataFrame(avg_income,region)
contingency_tab
```

```python
from scipy.stats import chi2_contingency
chi2, p_value, _, _ = chi2_contingency(contingency_tab)

print("Average Income level by region:")
print(avg_income)
print("")

print("Chi-Square Test Results:")
print("Chi-Square:", chi2)
print("p-value:", p_value)

print("Minimum average income by region:", avg_income.min())
print("Maximum average income by region:", avg_income.max())
```

Observation

• The contingency table shows the distribution of income group across different region. 

• Since the chi-square score t is -0.0913, this tells us that they are negatively correlated.

• And since the corresponding p-value is 1.0 it is greater than 0.05 (significance level), we conclude that the two variables are not statistically significant.

H1: Countries with higher population are more likely to have high remittance received

```python
x = df['population']
y = df['remittance_received']
pearsonr(x, y)
```

```python
#create a scatter plot
plt.figure(figsize=(12,5))
plt.scatter(x,y,c='purple')
plt.xlabel('Population')
plt.ylabel('Remittance received')
plt.title("Population vs. Remittance received")
```

Observation

• As population increases, the remittance received increases to some extent

• Since the correlation coefficient is 0.504, this tells us that they are moderately correlated.

• And since the corresponding p-value is 0.0 it is less than 0.05 (significance level), we conclude that the two variables are statistically significant.

Hypothesis testing factors:

Steps involved in hypothesis testing:
    1. Formulate your hypotheses
    2. Choosing a level of significance
    3. Identifying a statistical test
    4. Calculating the test statistic
    5. Calculating the p-value and comparing it with the level of signficance 
    6. Concluding - rejecting or not rejecting your null hypothesis
